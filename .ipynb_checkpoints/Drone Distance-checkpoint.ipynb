```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
dji_mavic_3 = {
    'focal_length': 24,
    'image_width': 1920,
    'image_height': 1080, 
    'sensor_width': 17.75, 
    'sensor_height': 11.83, 
}

def distance_estimation(bbox, altitude=50, alpha=20, cam_config=dji_mavic_3): 
    x_min = bbox['x1']
    y_min = bbox['y1']
    x_max = bbox['x2']
    y_max = bbox['y2']

    x_f = (x_min + x_max) / 2
    y_f = y_max
    
    c_x = int(cam_config['image_width'] / 2)
    c_y = int(cam_config['image_height'] / 2)

    f_x = (cam_config['focal_length'] / cam_config['sensor_width']) * cam_config['image_width'] # focal length in pixels
    
    d = math.sqrt((c_x - x_f)**2 + (c_y - y_f)**2) / f_x 
    theta = math.atan(d)
    alpha = math.radians(alpha)
    #Principle Point calculations 
    O_P = altitude * (1 / math.cos(alpha))
    O_F = 0
    
    
    if y_f > c_y: 
        O_F = altitude * (1 / math.cos(alpha - theta))
    else: 
        O_F = altitude * (1 / math.cos(alpha + theta))
    
    G_F = math.sqrt(O_F**2 - altitude**2)

    return O_F, G_F, theta 
```

```python
all_data = []
for alpha in range(0, 90, 10): 
    data = []
    for x in range(10, 1920, 10): 
        for y in range(10, 1080, 10): 
            x1 = x - 10
            y1 = y - 10
            x2 = x
            y2 = y 

            bbox = {
                'x1': x1, 
                'y1': y1, 
                'x2': x2, 
                'y2': y2
            }

            distance = distance_estimation(bbox=bbox, alpha=alpha)
            entry = [x1, y1, x2, y2, distance[1], distance[2]]
            data.append(entry)
    all_data.append(data)
```

```python
alph = 0
for data in all_data: 
    df = pd.DataFrame(data, columns=['x1', 'y1', 'x2', 'y2', 'distance', 'theta'])
    df = df.pivot('y2', 'x2', 'distance')
    ax = sns.heatmap(df)
    plt.title('Drone Distance Heatmap, alpha={}'.format(alph))
    plt.show()
    alph += 10
```
